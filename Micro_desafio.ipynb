```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("micro_desafio.csv")
```

```python
ds.head(23)
```

```python
ds.info()
```

```python
media_de_gols_por_partida = ds["gols_brasil"].mean()
print(f"A média de gols do Brasil por partida é: {media_de_gols_por_partida:.2f}")
```

```python
partidas = range(1, len(ds) + 1)
gols_brasil = ds["gols_brasil"]

plt.plot(partidas, gols_brasil, marker='.')

plt.title("Gols do Brasil por partida")
plt.xlabel("Partida")
plt.ylabel("Gols do Brasil")

plt.show()
```

```python
partidas = range(1, len(ds) + 1)
gols_brasil = ds["gols_brasil"]

plt.plot(partidas, gols_brasil, marker='.')
plt.axhline(media_de_gols_por_partida, linestyle="--", label=f"Média ({media_de_gols_por_partida:.2f})")

plt.title("Gols do Brasil por partida")
plt.xlabel("Partida")
plt.ylabel("Gols do Brasil")

plt.show()


partidas_acima_da_media = len(ds[ds["gols_brasil"] > media_de_gols_por_partida])
print(f"\nA média de gols é {media_de_gols_por_partida:.2f} e a quantidade de partidas acima da média são: {partidas_acima_da_media} partidas")
```

```python
times = ds["time_adversario"]
gols_adversario = ds["gols_adversario"]

plt.figure(figsize=(12,6))
plt.bar(times, gols_adversario)

plt.title("Gols marcados pelos adversários contra o Brasil")
plt.xlabel("Times adversários")
plt.ylabel("Gols dos Adversários")

plt.xticks(rotation=45)

plt.show()
```

```python
gols_por_adversario = ds.groupby("time_adversario")["gols_brasil"].sum()

plt.figure(figsize=(12,6))
plt.bar(gols_por_adversario.index, gols_por_adversario)

plt.title("Gols marcados pelo Brasil contra seus adversários")
plt.xlabel("Times adversários")
plt.ylabel("Gols do Brasil")

plt.xticks(rotation=45)

plt.show()
```
