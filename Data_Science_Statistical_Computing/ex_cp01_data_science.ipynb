```python
# Numero 26
# Pandas sob Comando — Enunciado

# Você recebeu um arquivo com registros acadêmicos (aluno, turma, nota, presença, entrega no prazo e projeto). Alguns campos podem estar ausentes.

# Calcule e guarde:
# 1) Quantos registros têm nota acima de 8 E presença acima de 85.
# 2) Quantos registros indicam entrega no prazo, desconsiderando registros sem nome do aluno.
# 3) Para cada turma, calcule a nota média (ignore notas ausentes). Guarde a maior dessas médias, arredondada para inteiro.
# 4) Encontre o registro com maior presença (ignore presenças ausentes) e guarde a nota desse registro (se a nota estiver ausente, use 0). Arredonde para inteiro.

# Some os quatro valores. Esse total é a resposta.

import pandas as pd

df = pd.read_csv("dataset_026.csv")
```

```python
df.info()
```

```python
df.head(62)
```

```python
# EX 01

quantidade_registros_nota_acima_presenca_acima_85 = df[(df['Nota'] > 8) & (df['Presenca_%'] > 85)].shape[0]
print(quantidade_registros_nota_acima_presenca_acima_85)
```

```python
# EX 02

quantidade_registros_entrega_no_prazo_sem_nome_aluno = df[(df['Entrega_no_prazo'] == 'Sim') & (df['Aluno'].notna())].shape[0]
print(quantidade_registros_entrega_no_prazo_sem_nome_aluno)
```

```python
# EX 03

nota_media_por_turma = df.groupby('Turma')['Nota'].mean()
maior_media_arredondada = int(round(nota_media_por_turma.max()))
print(maior_media_arredondada)
```

```python
# EX 04

# Encontre o índice do registro com maior presença
# idxmax() retorna o índice da primeira ocorrência do valor máximo, ignorando NaNs
idx_maior_presenca = df['Presenca_%'].idxmax()

# Obtenha a nota desse registro
nota_do_registro_maior_presenca = df.loc[idx_maior_presenca, 'Nota']

# Se a nota estiver ausente (NaN), use 0
if pd.isna(nota_do_registro_maior_presenca):
    nota_do_registro_maior_presenca = 0

# Arredonde para inteiro
nota_do_registro_maior_presenca = int(round(nota_do_registro_maior_presenca))

print(nota_do_registro_maior_presenca)
```

```python
idx_max_presenca = df['Presenca_%'].idxmax()
nota_alvo = df.loc[idx_max_presenca, 'Nota']
# Se a nota for NaN, v4 vira 0, senão arredonda
v4 = round(nota_alvo) if pd.notna(nota_alvo) else 0

print(v4)
```

```python
maior_media = df.groupby('Turma')['Nota'].mean().max()
v3 = round(maior_media)
print(v3)
```
